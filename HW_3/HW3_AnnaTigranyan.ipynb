```python
#Take data points where the absolute values of the ages are less than 100
df1 = df.loc[abs(df['age'])<100] 

fig, ax = plt.subplots(figsize = (8,5),dpi=100,facecolor="beige")
# Connection between age and popularity(scatter plot):


ax.scatter(x=df1['age'],y=df1['popularity'],color='cyan',edgecolors='green',
                linewidths=2,marker = 'p')

ax.set_title('Popularity by age',fontsize = 25, color = 'green', backgroundcolor = 'honeydew')# Add a title to the axes.
ax.set_ylabel('Popularity',color ='green',size = "15") # Add a y-label to the axes.
ax.set_xlabel('Age',color ='Green',size = "15") # Add a x-label to the axes.

# colering and linewidths of x and y axes 

for axis in ['bottom','left']:
    ax.spines[axis].set_linewidth(5)
    ax.spines[axis].set_color("indigo")
for axi in ['top','right']:  
    ax.spines[axi].set_color(None)
```

```python
# groupby houeses
groups = df[df['popularity']>0.6].groupby('house')

fig, ax = plt.subplots(figsize = (12,8))
ax.margins(0.09)

i = 0 # for color
for name, group in groups:
    # for each group plotting point which represents number of Dead Relations by age
    t = ax.plot(group.age, group.numDeadRelations, marker='o', linestyle='', ms=20, label=name, alpha = 0.8)
    t[0].set_linestyle('solid')
    i += 1 # changing color
#set legends    
ax.legend(loc = 'upper right', fontsize = 10, title = 'Houses')
#set labels
ax.set_xlabel('Age',fontsize = 20, color = 'r', backgroundcolor = 'bisque')
ax.set_ylabel('Number of Dead Relatives', fontsize = 20, color = 'r', backgroundcolor = 'bisque')
#set title
ax.set_title('Houses with maximum death',fontsize = 20, color = 'r', backgroundcolor = 'bisque')

#colorinf and linewidth of axes
for axis in ['top','bottom','left','right']:
    ax.spines[axis].set_linewidth(3)
    ax.spines[axis].set_color('orangered')

;
```

```python
# Pie chart

labels = 'NO', 'Yes'
x = df['isMarried'].value_counts()

fig, ax = plt.subplots(figsize = (5,5),dpi=100)

ax.pie(x, labels=labels, autopct='%1.0f%%',shadow=True, startangle=100,
       colors=['tomato','gold'],radius=0.5)

ax.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
ax.legend(['Yes','No'])
ax.set_title('Pie plot of "isMarried"', color = 'darkorange',fontsize = 20)
;
```

```python
# Pie chart

labels = 'NO', 'Yes'
x = df['isNoble'].value_counts()

fig, ax = plt.subplots(figsize = (5,5),dpi=100)
explode = (0.05, 0)
ax.pie(x, labels=labels,explode=explode,
       autopct='%1.0f%%',shadow=True, startangle=80,
       colors=['teal','orchid'],radius=1)

ax.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
ax.legend(['Yes','No'])
ax.set_title('Pie plot of "isNoble"', color = 'purple', backgroundcolor = 'honeydew',fontsize = 20)
;
```

```python
fig, ax = plt.subplots(figsize=(10, 5),dpi=100,facecolor='bisque')
box = ax.boxplot([df["popularity"], df1["alive"]], showfliers=False,
                 vert=False, patch_artist=True,
                medianprops={"color": "white", "linewidth": 1},
                boxprops={"facecolor": "pink", "edgecolor": "white",
                          "linewidth": 0.5});

# fill with colors
colors = ['gold', 'lime']
for bplot in box:
    for patch, color in zip(box['boxes'], colors):
        patch.set_facecolor(color)
        

    
ax.yaxis.grid(True)
ax.set_title('BoxPlots of "popularity" and "alive" ',fontsize = 20, color = 'r')

for axis in ['top','bottom','left','right']:
    ax.spines[axis].set_linewidth(3)
    ax.spines[axis].set_color('orangered')
```
