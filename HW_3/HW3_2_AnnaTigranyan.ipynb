```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv(r'character-deaths.csv')
```

```python
df.head()
```

```python
df.info()
```

```python

fig = plt.figure(figsize=(8,4), dpi=80)
ax = fig.add_axes([0,0,1,1])

a = sns.countplot(df["Gender"] , palette="PuRd")
# print count numbers in plot
for p in a.patches:
    a.annotate(f'\n{p.get_height()}', (p.get_x()+0.2, p.get_height()), va='top', color='black', size=22)   
# coloring, sizes an names of x and y labels    
ax.set_ylabel('Count',color ='indigo',size = "15") 
ax.set_xlabel('Gender(0-F, 1-M)',color ='indigo',size = "15") 
# setting linewidths and colors (axes)
for axis in ['top','bottom','left','right']:
    ax.spines[axis].set_linewidth(3)
    ax.spines[axis].set_color("indigo")
# setting title
plt.title('Countplot of Gender ',size=20, color = 'indigo', backgroundcolor = 'lavenderblush')
plt.show()
;
```

From a above we see that we have **157 Female** and **760 Male** characters in our data.

```python

fig = plt.figure(figsize=(8,4), dpi=80)
ax = fig.add_axes([0,0,1,1])

a = sns.countplot(df["Gender"] ,hue = df['Nobility'], palette="PuRd")
# print count numbers in plot
for p in a.patches:
    a.annotate(f'\n{p.get_height()}', (p.get_x()+0.2, p.get_height()),ha='center',va='top', color='black', size=22)   
# coloring, sizes an names of x and y labels    
ax.set_ylabel('Count',color ='indigo',size = "15") 
ax.set_xlabel('Gender(0-F, 1-M)',color ='indigo',size = "15") 
# setting linewidths and colors (axes)
for axis in ['top','bottom','left','right']:
    ax.spines[axis].set_linewidth(3)
    ax.spines[axis].set_color("indigo")
# setting title
plt.title('Countplot of Gender by Nobility',size=20, color = 'indigo', backgroundcolor = 'lavenderblush')
plt.show()
;
```

From a above countplot we see that we have:<br> 
<br>**157** Female characters of which<br> 1.**84** are **noble**<br> 2.**73** are **not**<br>
<br>**760** Female characters of which<br> 1.**73** are **noble**<br> 2.**346** are **414** 

```python
fig = plt.figure(figsize=(10,12))
ax = fig.add_axes([0,0,1,1])

a = sns.countplot(y = df["Book Intro Chapter"] ,color = 'blueviolet')

ax.set_ylabel('Book Intro Chapter',color ='indigo',size = "15") 
ax.set_xlabel('Count of characters',color ='indigo',size = "15") 

for axis in ['top','bottom','left','right']:
    ax.spines[axis].set_color("white")
    
plt.title("Countplot 'Book Intro Chapter'",size=20)
plt.show()
;
```

From above  we can see that the most  characters are introduced in:<br>
    1. **0.0** chapter <br>
    2. **2.0** chapter <br>
    3. **19.0** chapter <br>
    4. **38.0** chapter <br>

```python
fig = plt.figure(figsize=(8,4))
ax = fig.add_axes([0,0,1,1])

a = sns.countplot(df["Death Year"] ,palette="RdPu")

for p in a.patches:
    a.annotate(f'\n{p.get_height()}', (p.get_x()+0.2, p.get_height()), va='bottom', color='black', size=22)  

ax.set_ylabel('Count of characters',color ='indigo',size = "15") 
ax.set_xlabel('Death Year',color ='indigo',size = "15") 

for axis in ['top','bottom','left','right']:
    ax.spines[axis].set_color("None")
    
plt.title("Countplot of 'Death Year'",size=20,loc='left', color = 'indigo', backgroundcolor = 'lavenderblush')
plt.show()
;
```
